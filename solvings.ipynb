```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('tech-purchase.csv')
df.head(10)
```

```python
#a)more complicated info about NULL VALUES in dataframe
null_values = df.isnull().any(axis=1)
null_values
```

```python
##Show all of the null containing rows
df[null_values]
```

```python
df[~null_values]
```

```python
#b)DROP
#b)	Drop rows that contain any null values
df = df[~null_values]
df.head(15)
```

```python
    #2-TASK
#a)	Using our (cleaned) dataset, create a new column that would contain only the months from the Order Date column

only_numeric_months = df["Order Date"].str[:2]
df["Order Month"] = only_numeric_months
df.head(15)
```

```python
#b)	Convert the month numbers into characters (f.e: 1 into January, 2 into February, and so on).

    #here we drop the rows where during connecting several datasets besides 
    #the data can be the NAME of column as "Order Date" , so we have to drop such Words
df.drop(df[df["Order Month"]=="Or"].index, inplace=True)
df
```

```python
#create a dictionary with value of month number and as a key the NAME of month
    #set to 'Order month' column
number_to_month = {'01': 'January', '02': 'February', '03': 'March', '04': 'April', '05': 'May', '06': 'June', 
           '07': 'July', '08': 'August', '09': 'September', '10': 'October', '11': 'November', '12': 'December',}
df['Order Month'] = df['Order Month'].apply(lambda x: number_to_month[x])
df
```

```python
#c) Create a new column that would contain only the state names taken from the Purchase Address column.
df['States'] = df['Purchase Address'].str.split().str[-2]
df.head(10)
```

```python
#d)Convert the Quantity Ordered column’s values into an integer type and the Price Each column’s values into a floating type.
df['Quantity Ordered'] = df['Quantity Ordered'].astype(dtype=int, copy=True)
df['Price Each'] = df['Price Each'].astype(dtype=float,copy=True)
```

```python
#check datatype of "Quantity Ordered" column
df["Quantity Ordered"].head(3)
```

```python
#check datatype of "Price Each" column
df["Price Each"].head(3)
```

4.At what time of the day the highest amount of money is spent on sales?

```python
time = df["Order Date"].str.split().str[-1]
time
```

```python
df['Time'] = time #here we are taking TIME values among Order Dare
df.head(10)
```

```python
def mapping_fun(sample):
    h, m = int(sample[:2]), int(sample[3:])   # hours and minutes
    if h >= 6 and h < 12:
        return "morning"
    elif h >= 12 and h < 17:
        return "afternoon"
    elif h >= 17 and h < 21:
        return "evening"
    else:
        return "night"

time = time.apply(lambda x: mapping_fun(x))
```

```python
df["Time"] = time
df["Spent Money"] = df["Quantity Ordered"] * df["Price Each"]
df
```

```python
# On average at the evening it is spent the most
pd.pivot_table(df, values='Spent Money', index='Time')
```

```python
# If we consider only max numbers, then afternoon, morning and night have the same max result 3400
pd.pivot_table(df, values='Spent Money', index='Time', aggfunc=np.max)
```
